## 7. O Método do lugar das raízes

### 7.1 Introdução

- A estabilidade relativa e o desempenho transitório de um sistema de controle em malha fechada estão relacionados à localização das raízes da equação característica no plano $s$.
- Muitas vezes, é necessário ajustar parâmetros do sistema para obter posições adequadas das raízes.
- O método do lugar das raízes, introduzido por Evans em 1948, é uma técnica gráfica que esboça a trajetória das raízes no plano $s$ à medida que um parâmetro varia.
- Esse método permite medir a sensibilidade das raízes a variações nos parâmetros do sistema.
- Pode ser usado em conjunto com o critério de Routh–Hurwitz para obter informações qualitativas sobre a estabilidade e o desempenho do sistema.
- A técnica também é aplicável tanto para sistemas de malha simples quanto múltipla.
- Se as localizações das raízes forem insatisfatórias, ajustes nos parâmetros podem ser facilmente identificados a partir do lugar das raízes.


### 7.2 O conceito de lugar geométrico da raiz

- O desempenho dinâmico de um sistema de controle em malha fechada é descrito pela função de transferência em malha fechada:

$$ T(s) = \frac{Y(s)}{R(s)} = \frac{p(s)}{q(s)} $$

- As raízes da equação característica $q(s)$ determinam os modos de resposta do sistema.
- Para um sistema de malha única, a equação característica é dada por:

$$ 1 + KG(s) = 0 $$

onde $K$ é um parâmetro variável.

- As raízes características do sistema devem satisfazer a equação acima, onde as raízes estão no plano $s$.
- Reescrevendo em forma polar:

$$ |KG(s)| = 1 $$
$$ \angle KG(s) = 180^\circ + k360^\circ, \quad k = 0, \pm1, \pm2, \dots $$

- O **lugar das raízes** é o caminho traçado pelas raízes da equação característica no plano $s$ à medida que o parâmetro $K$ varia de zero ao infinito.

- Para um sistema de segunda ordem, a equação característica é:

$$ \Delta(s) = s^2 + 2s + K = s^2 + 2\zeta \omega_n s + \omega_n^2 = 0 $$

- As raízes, para $\zeta < 1$, são dadas por:

$$ s_{1,2} = -\zeta \omega_n \pm \omega_n \sqrt{1 - \zeta^2} $$

- A condição de magnitude é:

$$ |KG(s)| = \frac{K}{|s(s + 2)|} = 1 $$

- A condição de ângulo é:

$$ \angle KG(s) = \pm 180^\circ, \pm 540^\circ, \dots $$

- Para sistemas em malha múltipla, a equação característica pode ser expressa como:

$$ q(s) = \Delta(s) = 1 + F(s) = 0 $$

onde:

$$ F(s) = \frac{K(s + z_1)(s + z_2) \dots (s + z_M)}{(s + p_1)(s + p_2) \dots (s + p_n)} $$

- As condições de magnitude e ângulo são:

$$ |F(s)| = 1 $$
$$ \angle F(s) = 180^\circ + k360^\circ, \quad k \text{ é um inteiro} $$

- Para um sistema de segunda ordem, a equação característica pode ser reescrita como:

$$ 1 + \frac{as}{s^2 + K} = 0 $$

- A magnitude de $a$ é encontrada por:

$$ a = \frac{|s_1 - j2K| |s_1 + j2K|}{|s_1|} $$

- A localização dos polos no plano $s$ muda conforme $K$ varia, traçando o **lugar das raízes**.

- Finalmente, o esboço do lugar das raízes segue critérios sistemáticos que serão explorados na próxima seção.


### 7.3 O Procedimento do lugar geométrico da raiz

Vamos estudar o procedimento do Lugar das Raízes:  

As raízes da equação característica de um sistema fornecem informações valiosas sobre a resposta do sistema. Para localizar graficamente essas raízes no plano $s$, seguimos um procedimento ordenado de sete passos que facilita o esboço rápido do lugar das raízes.  

- **Passo 1:** Preparar o esboço do lugar das raízes. Comece escrevendo a equação característica como:  
$$ 1 + F(s) = 0. $$  

Rearranje a equação, se necessário, para que o parâmetro de interesse, $K$, apareça como fator multiplicativo na forma:  
$$ 1 + KP(s) = 0. $$  

Nos interessa determinar o lugar das raízes quando $K$ varia de $0$ a $\infty$. No entanto, na Seção 7.7, consideramos o caso em que $K$ varia de $-\infty$ a $0$.  

Fatorando $P(s)$, escrevemos o polinômio na forma de polos e zeros:  
$$ 1 + K \frac{\prod_{i=1}^{M}(s + z_i)}{\prod_{j=1}^{n}(s + p_j)} = 0. $$  

Localize os polos $-p_i$ e zeros $-z_i$ no plano $s$ com símbolos específicos. Por convenção:  
- "x" representa polos  
- "o" representa zeros  

Reescrevendo a Equação anterior:  
$$ \prod_{j=1}^{n}(s + p_j) + K \prod_{i=1}^{M}(s + z_i) = 0. $$  

Note que esta é outra forma de escrever a equação característica.  
- Quando $K = 0$, as raízes da equação característica são os polos de $P(s)$, pois:  
$$ \prod_{j=1}^{n}(s + p_j) = 0. $$  

A solução dessa equação fornece os valores de $s$ que coincidem com os polos de $P(s)$.  

- Por outro lado, quando $K \rightarrow \infty$, as raízes da equação característica tendem a ser os zeros de $P(s)$.  

Dividindo a equação por $K$:  
$$ \frac{1}{K} \prod_{j=1}^{n}(s + p_j) + \prod_{i=1}^{M}(s + z_i) = 0, $$  

que, à medida que $K \rightarrow \infty$, resulta em:  
$$ \prod_{i=1}^{M}(s + z_i) = 0. $$  

A solução desta equação fornece os valores de $s$ que coincidem com os zeros de $P(s)$.  

Portanto, concluímos que o lugar das raízes da equação característica $1 + KP(s) = 0$ começa nos polos de $P(s)$ e termina nos zeros de $P(s)$ à medida que $K$ cresce de zero ao infinito.  

Para a maioria das funções $P(s)$ que encontraremos, vários zeros de $P(s)$ estão no infinito no plano $s$, já que geralmente há mais polos do que zeros.  

Com $n$ polos e $M$ zeros, sendo $n > M$, há $n - M$ ramos do lugar das raízes que se aproximam dos $n - M$ zeros no infinito.  

- **Passo 2:** Localizar os segmentos do eixo real que pertencem ao lugar das raízes. O lugar das raízes no eixo real sempre se encontra à esquerda de um número ímpar de polos e zeros.  

Esse fato é determinado ao examinar o critério de ângulo, dado pela Equação (7.17).  

Esses dois passos iniciais para traçar um lugar das raízes serão ilustrados por um exemplo adequado.  


#### Exemplo 7.1: Sistema de Segunda Ordem

Um sistema de controle em malha fechada possui a equação característica:

$$
1 + G_c(s)G(s) = 1 + \frac{K(2s + 2)}{s^2 + 4s} = 0.
$$

#### Passo 1: Reescrevendo a equação característica

Podemos reescrevê-la como:

$$
1 + \frac{K(2s + 2)}{s(s + 4)} = 0.
$$

Para determinar o lugar das raízes para $0 \leq K < \infty$, localizamos os polos e zeros no eixo real.

- Polos: $s = 0$ e $s = -4$
- Zero: $s = -1$

#### Passo 2: Critério do Ângulo

O critério do ângulo é satisfeito no eixo real entre os pontos $0$ e $-2$:

- Ângulo do polo $s = 0$: $180^\circ$
- Ângulo do polo $s = -4$: $0^\circ$

O lugar das raízes começa nos polos e termina nos zeros, resultando no seguinte:

- Uma trajetória termina no zero real $s = -1$.
- A outra trajetória vai para o infinito negativo.

A fórmula para o ganho $K$ em uma raiz específica $s = s_1 = -1$ é dada por:

$$
\left| \frac{K (s_1 + 2)}{(s_1)(s_1 + 4)} \right| = 1
$$

Calculando $K$ para $s_1 = -1$:

$$
K = \frac{|-1| \cdot |-1 + 4|}{2 \cdot |-1 + 2|} = \frac{3}{2}
$$

Outro polo está localizado em $s = -6$.

#### Passo 3: Cálculo dos Assíntotas

O número de assíntotas $N$ é dado por:

$$
N = n - M
$$

onde:

- $n = 2$ (número de polos)
- $M = 1$ (número de zeros)

Logo, $N = 2 - 1 = 1$.

A posição das assíntotas é centrada em $s_A$:

$$
s_A = \frac{\sum \text{polos} - \sum \text{zeros}}{n - M}
$$

$$
s_A = \frac{(0 + (-4)) - (-2)}{2 - 1} = -1
$$

Os ângulos das assíntotas $\phi_A$ são dados por:

$$
\phi_A = \frac{(2k + 1)180^\circ}{n - M}, \quad k = 0, 1, 2, \dots, (n - M - 1)
$$

Substituindo:

- Para $k = 0$: $\phi_A = 90^\circ$
- Para $k = 1$: $\phi_A = 270^\circ$

Portanto, as assíntotas estão localizadas em $s_A = -1$ com ângulos de $90^\circ$ e $270^\circ$.

---

Esse exemplo demonstra como construir o lugar das raízes de um sistema de segunda ordem, localizando polos e zeros, determinando assíntotas e calculando o ganho $K$ em pontos específicos.


#### Exemplo 7.2: Sistema de Quarta Ordem

Um sistema de controle em malha fechada com realimentação unitária tem a seguinte equação característica:

$$
1 + G_c(s)G(s) = 1 + \frac{K(s + 1)}{(s + 2)(s + 4)} = 0.
$$

#### Passo 1: Localização dos Polos e Zeros

- Polos: $s = -2$ e $s = -4$
- Zero: $s = -1$

#### Passo 2: Assíntotas

O número de assíntotas $N$ é dado por:

$$
N = n - M = 4 - 1 = 3
$$

O ponto de interseção das assíntotas é:

$$
s_A = \frac{(-2) + (-4) - (-1)}{4 - 1} = -3
$$

Os ângulos das assíntotas são:

- $60^\circ$ ($k = 0$)
- $180^\circ$ ($k = 1$)
- $300^\circ$ ($k = 2$)

#### Passo 3: Ponto de Ruptura

O ponto de ruptura no eixo real ocorre quando:

$$
\frac{dK}{ds} = 0
$$

A função $p(s)$ é dada por:

$$
p(s) = - (s + 2)(s + 4) = - (s^2 + 6s + 8)
$$

Derivando $p(s)$:

$$
\frac{dp(s)}{ds} = - (2s + 6) = 0
$$

Resolvendo:

$$
s = -3
$$

Portanto, o ponto de ruptura ocorre em $s = -3$.

#### Resumo

- O lugar das raízes começa nos polos em $s = -2$ e $s = -4$.
- Três assíntotas se encontram em $s = -3$ com ângulos de $60^\circ$, $180^\circ$ e $300^\circ$.
- O ponto de ruptura no eixo real está em $s = -3$.

Esse exemplo demonstra a localização dos polos e zeros, a determinação das assíntotas e o cálculo analítico do ponto de ruptura, utilizando tanto métodos gráficos quanto diferenciais.


#### Exemplo 7.3 — Sistema de terceira ordem

O exemplo analisa um sistema de controle em malha fechada, cuja equação característica é:

$$
1 + G(s)H(s) = 1 + \frac{K(s + 1)}{s(s + 2)(s + 3)} = 0
$$

#### Principais pontos:
- O número de polos $n = 3$ e de zeros $M = 1$, resultando em **duas assíntotas** a $ \pm 90^\circ $ com centro em $s_A = -2$.
- Um **ponto de separação** ocorre entre $s = -2$ e $s = -3$. Para encontrá-lo, a equação característica é reorganizada para isolar $K$:

$$
K = - \frac{s(s + 2)(s + 3)}{s + 1}
$$

- Avaliando $p(s)$ numericamente, identifica-se o ponto de separação em $s = -2.46$.
- O **ângulo de partida** de polos complexos é determinado pela diferença entre os ângulos dos vetores dos outros polos e zeros, de acordo com o critério de ângulo:

$$
\theta_{\text{partida}} = 90^\circ - \theta_3
$$

- Para polos conjugados, o ângulo de partida de $-p_2$ é o oposto do de $-p_1$.
- O **ângulo de chegada** nos zeros também segue o critério de fase.

#### Passos resumidos para esboçar o lugar das raízes:
1. Determinar a equação característica.
2. Contar polos e zeros.
3. Localizar assíntotas.
4. Encontrar pontos de separação.
5. Calcular ângulos de partida e chegada.
6. Usar o critério de fase para raízes específicas.
7. Completar o esboço final do lugar das raízes.

Esses passos são resumidos na Tabela 7.2.


#### Exemplo 7.4 Sistema de Quarta Ordem

1. **Equação Característica**:
   A equação característica do sistema é dada por:

   $$
   1 + \frac{K}{s^4 + 12s^3 + 64s^2 + 128s} = 0
   $$

2. **Determinação dos polos**:
   Os polos são encontrados como raízes da equação:

   $$
   1 + \frac{K}{s_1s + 4j42s + 4 - j42} = 0
   $$

   O sistema não possui zeros finitos.

3. **Loci das raízes**:
   O número de polos $n$ é 4, então existem quatro loci. Eles são simétricos em relação ao eixo real. Um segmento do root locus existe entre $s = 0$ e $s = -4$.

4. **Ângulos das assíntotas**:
   Os ângulos das assíntotas são dados por:

   $$
   \theta_A = 12k + 12 \quad \text{com} \quad k = 0, 1, 2, 3

5. **Equação Característica Reescrita**:
   A equação característica reescrita é:

   $$
   s^4 + 12s^3 + 64s^2 + 128s + K = 0
   $$

   A matriz de Routh é:

   $$
   \begin{array}{|c|c|c|c|c|}
   s_4 & 1 & 64 & K \\
   s_3 & 12 & 128 & b_1 \\
   s_2 & b_1 & c_1 & K \\
   s_1 & c_1 & 0 &  \\
   s_0 & K &  &  \\
   \end{array}
   $$

   Com $b_1 = 53.33$ e $c_1 = 53.33$.

6. **Ponto de Quebra**:
   O ponto de quebra ocorre em $s \approx -1.577$.

7. **Ângulo de Partida**:
   O ângulo de partida em um polo complexo é:

   $$
   u_1 + 90^\circ + 90^\circ + u_3 = 180^\circ + k360^\circ
   $$

8. **Esboço Completo do Root Locus**:
   O gráfico completo do root locus é obtido visualmente, considerando os pontos de cruzamento no eixo imaginário e os efeitos da raiz dominante.

   A resposta transiente pode ser escrita como:

   y(t) = 1 + c_1 e^{-s_1 t} \sin(v_1 t + \phi_1) + c_2 e^{-s_2 t} \sin(v_2 t + \phi_2)
   $$

   Onde $s_1$ e $s_2$ são as raízes dominantes.


### 7.4 Projeto de Parâmetros pelo Método do Root Locus

1. **Objetivo do Método**:
   O método do root locus foi originalmente desenvolvido para determinar o locus das raízes da equação característica à medida que o ganho do sistema, $K$, varia de zero a infinito. No entanto, ele também pode ser usado para investigar o efeito de outros parâmetros do sistema.

2. **Equação Característica**:
   A equação característica do sistema é dada por:

   $$
   1 + F_1 s^2 = 0
   $$

   O método padrão do root locus é aplicado para investigar os efeitos de parâmetros do sistema.

3. **Investigação de Parâmetros Múltiplos**:
   O método root locus é essencialmente um método de um único parâmetro, mas pode ser estendido para investigar dois ou mais parâmetros. Isso é feito isolando cada parâmetro e aplicando o método do root locus.

4. **Equação Característica Geral**:
   Para um sistema dinâmico, a equação característica pode ser escrita como:

   $$
   a_n s^n + a_{n-1} s^{n-1} + \dots + a_1 s + a_0 = 0
   $$

   O efeito da variação de um parâmetro, como $a_1$, pode ser investigado a partir da equação de root locus:

   $$
   1 + \frac{a_1 s}{a_n s^n + a_{n-1} s^{n-1} + \dots + a_0} = 0
   $$

5. **Isolando Parâmetros**:
   Caso o parâmetro de interesse, $a$, não apareça apenas como um coeficiente, ele pode ser isolado e a equação reescrita no formato de root locus.

6. **Exemplo de Equação de Terceira Ordem**:
   Considere a equação de terceira ordem:

   $$
   s^3 + 13 + a_2 s^2 + 3s + 6 = 0
   $$

   Para estudar o efeito do parâmetro $a$, isolamos o parâmetro e reescrevemos a equação como:

   $$
   1 + a_1 s^2 = 0
   $$

7. **Investigação de Dois Parâmetros**:
   Para investigar dois parâmetros, $a$ e $b$, o processo é repetido duas vezes. Primeiro, investigamos o efeito de $a$ e, em seguida, o efeito de $b$.

8. **Exemplo de Equação com Dois Parâmetros**:
   Uma equação característica com dois parâmetros seria:

   $$
   s^3 + s^2 + b s + a = 0
   $$

   O efeito de $b$ pode ser investigado variando de zero a infinito a partir da equação:

   $$
   1 + b s = 0
   $$

9. **Passos para Avaliação**:
   Primeiro, avaliamos o efeito de $a$ utilizando a equação $s^3 + s^2 + a = 0$. Em seguida, avaliamos o efeito de $b$ utilizando a equação com $a$ fixo.

10. **Limitações**:
   Uma limitação desse método é que nem sempre será possível obter uma equação característica linear no parâmetro considerado.

11. **Exemplo Prático**:
   O método é ilustrado com um exemplo de design, onde a locação das raízes é obtida para $a$ e depois para $b$, resultando em uma seleção de parâmetros adequados para o design do sistema.



#### Exemplo 7.5 Controle de Cabeça de Soldagem

1. **Objetivo do Sistema**:
   Um sistema de controle de feedback deve ser projetado para a posição de uma cabeça de soldagem em um carro. As especificações do controle incluem:

   - Erro de estado estacionário para uma entrada rampante ($e_{ss}$) ≤ 0.35, com inclinação de entrada.
   - A razão de amortecimento das raízes dominantes ($\zeta$) ≥ 0.707.
   - Tempo de assentamento ($T_s$) dentro de 2% do valor final, $T_s$ ≤ 3s.

2. **Estrutura do Sistema de Controle**:
   O sistema de feedback é composto por um ganho amplificador $K_1$ e um ganho de feedback derivativo $K_2$, que precisam ser selecionados para satisfazer as especificações.

3. **Erro de Estado Estacionário**:
   O erro de estado estacionário é dado por:

   $$
   e_{ss} = \lim_{t \to \infty} e(t) = \lim_{s \to 0} s E(s) = \frac{R(s)}{1 + G_2(s)} \quad \Rightarrow \quad e_{ss} = \frac{2 + K_1 K_2}{K_1} \leq 0.35.
   $$

   Para atender a essa exigência, $K_2$ é escolhido pequeno.

4. **Razão de Amortecimento e Localização das Raízes**:
   A razão de amortecimento exige que as raízes do sistema fechado fiquem abaixo da linha de 45° no plano complexo esquerdo. A condição de tempo de assentamento é dada por:

   $$
   T_s = 4 / \sigma \quad \Rightarrow \quad \sigma \geq 4 / 3 \quad \text{(onde $\sigma$ é a parte real das raízes dominantes)}.
   $$

   As raízes devem estar à esquerda da linha $s = -4/3$.

5. **Equação Característica**:
   A equação característica para o sistema é:

   $$
   s^2 + 2s + bs + a = 0.
   $$

   O locus das raízes, variando $a = K_1$ (com $b = 0$), é dado por:

   $$
   1 + \frac{a}{s^2 + 2s + 2} = 0.
   $$

   Para $K_1 = 20$, as raízes são $s = -1 \pm j4.36$.

6. **Efeito de Variação de $b = 20K_2$**:
   A equação do locus para o efeito de $b$ é dada por:

   $$
   1 + \frac{b}{s^2 + 2s + 20} = 0.
   $$

   A raiz de amortecimento $\zeta = 0.707$ é alcançada para $b = 4.3$, ou $K_2 = 0.215$, com a parte real das raízes $s = -3.15$. O tempo de assentamento resultante é $T_s = 1.27$ s, que é menor que a especificação de $T_s = 3$ s.

7. **Método do Root Locus com Dois Parâmetros**:
   O método do root locus pode ser estendido para dois parâmetros. A equação característica para dois parâmetros $a$ e $b$ é:

   $$
   s^3 + 3s^2 + 2s + bs + a = 0.
   $$

   O locus das raízes em função de $a$ (com $b = 0$) é dado por:

   $$
   1 + \frac{a}{s^3 + 3s^2 + 2s + 2} = 0.
   $$

   O locus das raízes em função de $b$ (com $a$ fixo) é dado por:

   $$
   1 + \frac{b}{s^3 + 3s^2 + 2s + a} = 0.
   $$

   A ilustração mostra o efeito de $a$ e $b$ nas raízes do sistema, com o método de loci de raízes gerando uma família de loci chamados "contornos de raízes", mostrando o impacto de $a$ e $b$.



### 7.5 Sensibilidade e o Locus das Raízes

#### **Introdução**:
   O feedback negativo em sistemas de controle é usado para reduzir os efeitos das variações de parâmetros. A sensibilidade do sistema à variação de parâmetros pode ser medida através da **sensibilidade logarítmica** proposta por Bode, dada por:

   $$
   S_K/T = \frac{d \ln T}{d \ln K} = \frac{d T}{d K}.
   $$

   Onde $T(s)$ é a função de transferência do sistema e $K$ é o parâmetro de interesse.

#### **Sensibilidade das Raízes**:
   A sensibilidade das raízes de um sistema pode ser definida como a variação da posição das raízes em relação à mudança do parâmetro $K$:

   $$
   S_K/r_i = \frac{d r_i}{d \ln K} = \frac{d r_i}{d K},
   $$

   Onde $r_i$ representa a $i$-ésima raiz do sistema. A função de transferência é dada por:

   $$
   T(s) = \frac{K_1}{\prod_{j=1}^{M} (s + z_j) \prod_{i=1}^{n} (s + r_i)}.
   $$

#### **Relação entre Sensibilidade Logarítmica e das Raízes**:
   A sensibilidade das raízes pode ser relacionada à sensibilidade logarítmica através da fórmula:

   $$
   S_T/K = \frac{d \ln K}{d \ln K} - \sum_{i=1}^{n} \frac{d r_i}{d \ln K} \frac{1}{s + r_i}.
   $$

   Para sistemas onde os zeros de $T(s)$ não dependem de $K$, temos que:

   $$
   \frac{d z_j}{d \ln K} = 0.
   $$

#### **Sensibilidade Logarítmica**:
   Quando o ganho do sistema não depende de $K$, a sensibilidade logarítmica é dada por:

   $$
   S_T/K = - \sum_{i=1}^{n} S_K/r_i \frac{1}{s + r_i}.
   $$

#### **Avaliação da Sensibilidade das Raízes**:
   A sensibilidade das raízes pode ser avaliada utilizando o método do locus das raízes. Para isso, variamos o parâmetro $K$ por um valor pequeno $\Delta K$ e observamos a mudança na raiz $r_i$, que se torna $r_i + \Delta r_i$. A sensibilidade da raiz é dada por:

   $$
   S_K/r_i \approx \frac{\Delta r_i}{\Delta K}.
   $$

   Esse é um valor aproximado, que se aproxima do valor real à medida que $\Delta K \to 0$.



#### Exemplo 7.6: Sensibilidade das Raízes de um Sistema de Controle

Este exemplo discute a avaliação da sensibilidade das raízes em um sistema de controle com o seguinte equação característica:

$$ 1 + \frac{K}{s^2 + bs} = 0 $$

ou, de forma equivalente:

$$ s^2 + bs + K = 0 $$

onde o ganho $ K $ é o parâmetro de interesse. O efeito das mudanças nos parâmetros (por exemplo, $ a = a_0 + \Delta a $ e $ b = b_0 + \Delta b $) é analisado para diferentes valores dos parâmetros.

#### Caso de Variação de Ganho:
- Os valores nominais são $ b_0 = 1 $ e $ K = a_0 = 0.5 $.
- O local das raízes do sistema é obtido, mostrando as raízes em $ -r_1 = -0.5 + j0.5 $ e $ -r_2 = -nr_1 $.

Quando o ganho $ a $ muda 20% (ou seja, $ \Delta a = \pm 0.1 $), as localizações das raízes mudam:
- Para $ K = 0.6 $, $ r_1 = -0.5 + j0.59 $, e a mudança é $ \Delta r_1 = +j0.09 $.
- Para $ K = 0.4 $, $ r_1 = -0.5 + j0.387 $, e a mudança é $ \Delta r_1 = -j0.11 $.

A sensibilidade das raízes para as mudanças positivas e negativas no ganho é calculada:
- $ S_{K+ r_1} = j0.45 $ para $ \Delta a = +0.1 $.
- $ S_{K- r_1} = -j0.55 $ para $ \Delta a = -0.1 $.

Para mudanças infinitesimais em $ K $, a sensibilidade é igual para aumentos ou diminuições no ganho, com o ângulo indicando a direção de movimento da raiz.

#### Caso de Variação de Pólos:
- Para uma variação em $ b = b_0 + \Delta b $, a equação característica torna-se:

$$ s^2 + s + \Delta b s + K = 0 $$

O local das raízes muda dependendo do valor de $ \Delta b $, com efeitos diferentes para mudanças positivas e negativas. Por exemplo, para uma mudança de 20% em $ b $ ($ \Delta b = \pm 0.20 $), a sensibilidade das raízes é:

- $ S_{b+ r_1} = 0.80 \angle -128^\circ $ para mudança positiva em $ b $.
- $ S_{b- r_1} = 0.625 \angle 39^\circ $ para mudança negativa em $ b $.

#### Conclusão:
- As medidas de sensibilidade $ S_{K+ r_1} $ e $ S_{b+ r_1} $ indicam como o sistema é sensível às mudanças nos parâmetros $ K $ e $ b $, respectivamente.
- A sensibilidade das raízes ajuda a comparar os efeitos das variações dos parâmetros nas localizações das raízes, que correspondem ao desempenho do sistema.
- Embora a sensibilidade das raízes seja uma medida útil, ela é limitada pelo número de cálculos necessários e pela dependência das localizações das raízes para representar o desempenho do sistema.

Esta abordagem oferece percepções sobre o efeito das variações dos parâmetros e é especialmente útil para avaliar os requisitos de tolerância para diferentes parâmetros.


### 7.6 Controladores PID

O controlador PID (Proporcional, Integral, Derivativo) é amplamente utilizado no controle de processos industriais. Sua função de transferência é dada por:

$$ G_c(s) = K_p + \frac{K_I}{s} + K_Ds $$

O controlador PID combina três termos: proporcional ($K_p$), integral ($K_I$), e derivativo ($K_D$). Ao ajustar os parâmetros, podemos ter diferentes versões do controlador:
- **PI**: Quando $K_D = 0$, o controlador é apenas proporcional e integral.
- **PD**: Quando $K_I = 0$, o controlador é apenas proporcional e derivativo.

O controlador PID pode ser visto como a cascata de dois controladores: PI e PD. A função de transferência resultante é:

$$ G_c(s) = (K_P + \frac{K_I}{s})(K_P + K_Ds) $$

O controlador PID também pode ser representado como uma equação quadrática com um pólo na origem e dois zeros.

Em sistemas com controladores PID, a escolha dos parâmetros ($K_p$, $K_I$, $K_D$) é crucial para o desempenho do sistema. Existem métodos de sintonia, como o método de Ziegler–Nichols, que ajusta os ganhos com base nas respostas de entrada do sistema. O método manual envolve ajustes de $K_p$, $K_I$, e $K_D$ até que o sistema atinja a estabilidade desejada.

Para um desempenho ideal, a sintonia pode ser feita empiricamente ou com base em simulações, e ajustes finos são feitos para otimizar a resposta do sistema.

O controlador PID é popular devido à sua simplicidade funcional e bom desempenho em uma ampla gama de condições operacionais.


#### Exemplo 7.7: Sintonia Manual do PID

Neste exemplo, a sintonia manual de um controlador PID é feita para um sistema fechado com parâmetros dados. O processo começa com a configuração de $K_I = 0$ e $K_D = 0$, e a partir daí o ganho proporcional $K_P$ é aumentado até que o sistema apresente oscilações sustentadas.

1. **Ajuste Inicial**:
   - Quando $K_P = 885.5$, ocorre uma oscilação sustentada com amplitude $A = 1.9$ e período $P = 0.83$ s.
   - O locus das raízes mostra que os pólos do sistema estão localizados em $s = \pm 7.5j$.

2. **Redução de $K_P$**:
   - Para obter uma resposta com uma redução de amplitude de aproximadamente um quarto, $K_P$ é reduzido pela metade para $K_P = 442.75$ e ajustado para $K_P = 370$.
   - O locus das raízes para $K_P = 370$ mostra que, ao aumentar $K_D$, os pólos complexos se movem para a esquerda, aumentando o fator de amortecimento e diminuindo o overshoot.

3. **Ajuste de $K_I$**:
   - Ao ajustar $K_I$, os pólos complexos se movem para a direita, diminuindo o fator de amortecimento e aumentando o overshoot. Quando $K_I = 778.2$, o sistema se torna marginalmente estável.

4. **Resultado Final**:
   - Os valores escolhidos para atender às especificações de overshoot e tempo de assentamento são: $K_P = 370$, $K_D = 60$, e $K_I = 100$.
   - A resposta ao degrau é satisfatória, com $T_s = 2.4$ s e $P.O. = 12.8$.

#### Métodos de Sintonia Ziegler-Nichols

Os métodos de sintonia Ziegler-Nichols, publicados em 1942, visam otimizar a resposta do sistema sem oscilações excessivas e com boa rejeição a distúrbios. Existem duas abordagens principais:
1. **Método de laço fechado**: Aumenta-se o ganho proporcional $K_P$ até que o sistema entre em oscilação sustentada. O ganho máximo é chamado de ganho último ($K_U$) e o período de oscilação é o período último ($P_U$). Com esses valores, os ganhos do PID são calculados.
2. **Método de laço aberto**: Baseado em curvas de reação, sem a necessidade de modelos precisos do processo.

A sintonia de Ziegler-Nichols é prática, mas pode não funcionar com todos os processos, sendo ideal para obter um design inicial que será refinado posteriormente.


#### Exemplo 7.8: Sintonia PID Ziegler–Nichols em Laço Fechado

Neste exemplo, os ganhos $K_P$, $K_D$ e $K_I$ são calculados usando as fórmulas do método de sintonia Ziegler–Nichols.

1. **Cálculo dos Ganhos**:
   - A partir de $K_U = 885.5$ e $T_U = 0.83$ s, os ganhos são calculados como:
     - $K_P = 0.6K_U = 531.3$
     - $K_I = 1.2K_U/T_U = 1280.2$
     - $K_D = 0.6K_U T_U / 8 = 55.1$

2. **Comparação com Sintonia Manual**:
   - A resposta ao degrau dos controladores com sintonia manual e Ziegler–Nichols apresenta tempos de assentamento semelhantes.
   - No entanto, o controlador sintonizado manualmente tem um overshoot menor, pois o método Ziegler–Nichols prioriza a rejeição de distúrbios em vez da resposta ao input.

3. **Desempenho de Rejeição de Distúrbios**:
   - O desempenho do controlador PID Ziegler–Nichols em relação a distúrbios é superior ao do controlador manualmente sintonizado, como mostrado na Figura 7.34.

4. **Método Ziegler–Nichols em Laço Aberto**:
   - O método Ziegler–Nichols em laço aberto utiliza uma curva de reação, obtida ao retirar o controlador do laço e aplicar um degrau de entrada.
   - Este método é comumente usado em controle de processos, sendo útil para sistemas de primeira ordem com atraso de transporte.

5. **Curva de Reação e Parâmetros**:
   - A curva de reação é caracterizada pelo atraso de transporte $\Delta T$ e a taxa de reação $R$. A equação do sistema pode ser aproximada como:
     $$ G_1(s) = \frac{M}{s + p} e^{-\Delta T s} $$
   - Onde $M$ é a magnitude da resposta, $\Delta T$ é o atraso e $p$ está relacionado à inclinação da curva.

6. **Cálculo dos Ganhos**:
   - Após estimar os parâmetros $\Delta T$, $M$, e $p$, a taxa de reação $R$ é calculada, e os ganhos PID são obtidos como mostrado na Tabela 7.6.

7. **Controle Proporcional e Proporcional-Integral**:
   - O método de sintonia Ziegler–Nichols também pode ser usado para projetar controladores proporcionais ou proporcionais-integral.



#### Exemplo 7.9: Sintonia Ziegler–Nichols PI em Laço Aberto

Neste exemplo, a curva de reação é usada para estimar o atraso de transporte $\Delta T = 0.1$ s e a taxa de reação $R = 0.8$. A sintonia do controlador PI usando o método Ziegler–Nichols resulta nos seguintes ganhos:

1. **Cálculo dos Ganhos**:
   - $K_P = 0.9 \frac{R}{\Delta T} = 11.25$
   - $K_I = 0.27 \frac{R}{(\Delta T)^2} = 33.75$

2. **Resposta ao Degrau do Sistema**:

   - Como o controlador é PI, o erro em regime permanente é zero, como esperado.

3. **Considerações sobre a Sintonia**:
   - As abordagens de sintonia manual e Ziegler–Nichols fornecem etapas estruturadas para projetar ganhos PID, mas podem não levar sempre ao desempenho desejado no laço fechado.
   - Esses métodos devem ser considerados como primeiros passos no processo de iteração do design, sendo importante conhecer diferentes abordagens de sintonia.

### 7.7 Raiz com Ganho Negativo

A **raiz do local** descreve a variação das raízes da equação característica à medida que um parâmetro de interesse varia. Para um sistema de controle com feedback unitário negativo, a equação característica é dada por:

$$ 1 + KG_1(s) = 0 $$

Onde $K$ é o parâmetro de interesse. Em situações com $K > 0$, a técnica padrão de esboço do **root locus** é aplicada. No entanto, em casos com **ganho negativo** ($-∞ \leq K \leq 0$), é necessário adaptar a metodologia.

1. **Equação Reorganizada**:
   A equação característica é reorganizada para:
   
   $$ G_1(s) = -\frac{1}{K} $$

   Com $K$ sendo negativo, a condição de magnitude e fase é:

   $$ |KG_1(s)| = 1 \quad \text{e} \quad \text{Fase}(KG_1(s)) = 0^\circ + k360^\circ $$

   Onde $k$ é um número inteiro.

2. **Modificações no Esboço**:
   As condições de fase para o caso de ganho negativo são diferentes das condições usuais e resultam em mudanças importantes no processo de esboço da raiz do local. A técnica exige ajustes nos passos tradicionais de esboço de raiz do local descritos em Seção 7.3.


#### Exemplo 7.10 Raiz do Local com Ganho Negativo

Considerando o sistema com a função de transferência do loop:

$$ L(s) = KG_1(s) = K \frac{s - 20}{s^2 + 5s - 50} $$

A equação característica é:

$$ 1 + K \frac{s - 20}{s^2 + 5s - 50} = 0 $$

O esboço do **root locus** para valores negativos de $K$ mostra que a estabilidade ocorre para $-5.0 \leq K \leq -2.5$, permitindo estabilizar o sistema apenas com **ganho negativo**.

#### Procedimento para Esboçar o Root Locus com Ganho Negativo

1. **Preparação do Esboço**:
   A equação característica é reorganizada para $1 + KP_1(s) = 0$. O objetivo é determinar o locus das raízes para $K \in (-∞, 0)$.

2. **Localização dos Segmentos no Eixo Real**:
   O **root locus** no eixo real ocorre em seções à esquerda de um número par de polos e zeros.

3. **Assíntotas**:
   Para $n > M$, temos assíntotas que vão para os zeros no infinito, com um ponto central $s_A$ e ângulos $\phi_A$ dados pelas equações:

   $$ s_A = \frac{\sum \text{polos} - \sum \text{zeros}}{n - M} $$

   $$ \phi_A = \frac{2k + 1}{n - M} \cdot 360^\circ $$

4. **Cruzamento do Eixo Imaginário**:
   A condição de Routh–Hurwitz é usada para determinar onde o locus cruza o eixo imaginário.

5. **Ponto de Separação**:
   O ponto de separação no eixo real pode ser determinado analiticamente ou graficamente, resolvendo a equação do locus.

6. **Ângulo de Partida e Chegada**:
   O ângulo de partida de um polo e o ângulo de chegada a um zero são calculados usando o critério de fase.

7. **Conclusão do Esboço**:
   O esboço final é completado com todos os segmentos não cobertos nos passos anteriores.

### 7.9 Local da raiz usando software de design de controle

O método do root locus pode ser usado manualmente para esboçar uma aproximação do gráfico. No entanto, o uso de software de design de controle permite gerar um gráfico de root locus preciso. Apesar disso, é fundamental entender os conceitos por trás do método de root locus, que são ensinados através dos passos manuais.

A seção descreve como gerar um gráfico de root locus utilizando software de controle. Além disso, discute a conexão entre a expansão de frações parciais, polos dominantes e a resposta do sistema em malha fechada, abordando também a sensibilidade dos polos.

Funções abordadas:
- `rlocus`: gera o gráfico de root locus.
- `rlocfind`: encontra valores específicos do gráfico de root locus.
- `residue`: realiza a expansão em frações parciais.

#### Passos para gerar o gráfico de root locus:
1. Obtenha a equação característica no formato necessário.
2. Utilize a função `rlocus` para gerar o gráfico.

No exemplo apresentado, à medida que o parâmetro de ganho $K$ aumenta, dois ramos do root locus se afastam do eixo real, indicando a presença de raízes complexas. A função `rlocfind` pode ser usada para determinar o valor de $K$ correspondente a essas raízes.

#### Exemplo de expansão de frações parciais:
A função `residue` é utilizada para expandir a função de transferência fechada, resultando na seguinte expansão:
$$ Y(s) = \frac{-1.3786 + j1.7010}{s + 2.0505 + j4.3228} + \frac{-1.3786 - j1.7010}{s + 2.0505 - j4.3228} + \frac{-0.2429}{s + 0.8989} $$

#### Cálculo do tempo de assentamento:
Usando o gráfico de resposta ao degrau, o tempo de assentamento é determinado. O valor calculado de 1,95 segundos é uma boa aproximação do valor real de 1,6 segundos.

#### Impacto dos zeros na resposta transitória:
A proximidade do zero em $s = -1$ com o polo em $s = -0.8989$ reduz o impacto desse polo na resposta transitória. O overshoot previsto para 60% é ajustado para 50% no gráfico real.

Esses cálculos e funções ajudam a entender o impacto dos polos e zeros na dinâmica do sistema e a estimar o comportamento da resposta transitória.


### 7.10 Exemplo de Projeto Sequencial: Sistema de Leitura de Disco

Neste capítulo, utilizamos o controlador PID para obter uma resposta desejável usando feedback de velocidade. O processo envolve modelar o sistema, selecionar um controlador e, finalmente, otimizar os parâmetros e analisar o desempenho. O método do root locus é usado na seleção dos parâmetros do controlador.

#### Seleção do Controlador
O controlador PID é dado por:
$$ G_c(s) = K_P + \frac{K_I}{s} + K_D s $$

Como o modelo do processo já possui uma integração, definimos $K_I = 0$, resultando em um controlador PD:
$$ G_c(s) = K_P + K_D s $$

Nosso objetivo é selecionar $K_P$ e $K_D$ para atender às especificações do sistema. A função de transferência em malha fechada do sistema é:
$$ T(s) = \frac{G_c(s) G_1(s) G_2(s)}{1 + G_c(s) G_1(s) G_2(s)} $$

#### Uso do Root Locus
A equação para o produto $G_c(s) G_1(s) G_2(s)$ é:
$$ G_c(s) G_1(s) G_2(s) = \frac{5000 K_P + K_D s}{s^2 + 2021 s + 10002} $$

Definimos $z = \frac{K_P}{K_D}$ para selecionar a localização do zero $z$. Com isso, podemos esboçar o gráfico do root locus como função de $K_D$. Ao escolher $z = 1$, obtemos:
$$ G_c(s) G_1(s) G_2(s) = \frac{5000 K_D (s + 1)}{s^2 + 2021 s + 10002} $$

#### Geração do Root Locus
O gráfico de root locus gerado por computador nos permite determinar as raízes para diferentes valores de $K_D$. Quando $K_D = 91.3$, obtemos as raízes mostradas no gráfico.

#### Análise de Desempenho
A resposta do sistema é analisada, e os tempos de assentamento são medidos. O sistema atinge 97% do valor final rapidamente e depois tende a se aproximar muito lentamente do valor final, levando um tempo de assentamento de 20 ms para atingir o valor final "praticamente". O sistema atende a todas as especificações.
